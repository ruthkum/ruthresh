```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Creating a sample dataframe
data = {
    'reviews[519].comments': ['Great product', 'Not bad', 'Could be better'],
    'other_column1': [1, 2, 3],
    'other_column2': [4, 5, 6],
    'url_column': ['https://www.airbnb.com/rooms/5739687', 'https://www.airbnb.com/rooms/1234567', 'https://www.airbnb.com/rooms/7654321']
}
df = pd.DataFrame(data)

# Dropping the column 'url_column' to create X
X = df.drop(["url_column"], axis=1)

# Encoding the target variable
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(df["reviews[519].comments"])

# Dropping the encoded target column from X
X = X.drop(["reviews[519].comments"], axis=1)

# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Define categorical features
categorical_features = []  # If you have any categorical features, list them here

# Create a preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(), categorical_features)
    ],
    remainder='passthrough'
)

# Create the full pipeline with preprocessing and the model
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Fit the model
pipeline.fit(X_train, y_train)

# Evaluate the model
print("Model trained successfully")
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load the dataset with caching
@st.cache_data
def load_data(file_path):
    df = pd.read_csv(file_path, encoding='ISO-8859-1', on_bad_lines='skip')
    cols_to_numeric = ['minimum_nights', 'accommodates', 'bedrooms', 'beds', 'number_of_reviews', 'maximum_nights', 'bathrooms']
    for col in cols_to_numeric:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    return df

# Load dataset from CSV file
file_path = '/content/sample_airbnb.listingsAndReviews (1).csv'
df = load_data(file_path)

# Generate questions, answers, and corresponding column names
questions_answers = [
    ("What is the sum of the minimum nights required across all listings?", df['minimum_nights'].sum(), 'minimum_nights'),
    ("What is the maximum number of accommodates in the listings?", df['accommodates'].max(), 'accommodates'),
    ("How many listings have more than 1 bedroom?", (df['bedrooms'] > 1).sum(), 'bedrooms'),
    ("What is the average number of beds per listing?", df['beds'].mean(), 'beds'),
    ("What is the total number of reviews for all listings?", df['number_of_reviews'].sum(), 'number_of_reviews'),
    ("What is the maximum number of nights allowed across all listings?", df['maximum_nights'].max(), 'maximum_nights'),
    ("What is the average number of bathrooms per listing?", df['bathrooms'].mean(), 'bathrooms'),
    ("How many listings have a flexible cancellation policy?", (df['cancellation_policy'] == 'Flexible').sum(), 'cancellation_policy'),
    ("What is the minimum number of accommodates in the listings?", df['accommodates'].min(), 'accommodates'),
    ("How many listings have exactly 1 bedroom?", (df['bedrooms'] == 1).sum(), 'bedrooms')
]

# Create DataFrame with questions, answers, and visualization types
qa_df = pd.DataFrame(questions_answers, columns=['Question', 'Answer', 'Column'])

# Streamlit App
st.title('Airbnb Listings Analysis')

# Select a question
question = st.selectbox('Select a question to view its answer and visualization:', qa_df['Question'])

# Display the answer
answer = qa_df[qa_df['Question'] == question]['Answer'].values[0]
column = qa_df[qa_df['Question'] == question]['Column'].values[0]
st.write(f'**Answer:** {answer}')

# Buttons for different types of visualizations
if st.button('Show Bar Chart'):
    if column in df.select_dtypes(include=['number']).columns:
        st.bar_chart(df[column])

if st.button('Show Line Chart'):
    if column in df.select_dtypes(include=['number']).columns:
        st.line_chart(df[column])

if st.button('Show Scatter Plot'):
    numeric_cols = df.select_dtypes(include=['number']).columns
    if len(numeric_cols) >= 2:
        plt.figure(figsize=(10, 6))
        sns.scatterplot(data=df, x=numeric_cols[0], y=numeric_cols[1])
        st.pyplot(plt.gcf())

if st.button('Show Histogram'):
    if column in df.select_dtypes(include=['number']).columns:
        plt.figure(figsize=(10, 6))
        df[column].plot(kind='hist', bins=20)
        plt.xlabel(column)
        plt.title(f'Histogram of {column}')
        st.pyplot(plt.gcf())

if st.button('Show Pie Chart'):
    plt.figure(figsize=(10, 6))
    if column == 'cancellation_policy':
        df[column].value_counts().plot.pie(
            autopct='%1.1f%%',
            startangle=90,
            labeldistance=1.5,
            pctdistance=0.85,
            textprops={'fontsize': 12}
        )
        plt.ylabel('')
        plt.title(f'Pie Chart of {column}')
    elif column in df.select_dtypes(include=['number']).columns:
        categorized_data = pd.cut(df[column], bins=10).value_counts()
        categorized_data.plot.pie(
            autopct='%1.1f%%',
            startangle=90,
            labeldistance=1.5,
            pctdistance=0.85,
            textprops={'fontsize': 12}
        )
        plt.ylabel('')
        plt.title(f'Pie Chart of {column}')
    st.pyplot(plt.gcf())
```
